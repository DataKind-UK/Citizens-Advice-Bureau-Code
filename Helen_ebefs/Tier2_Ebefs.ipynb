### Notebook for plotting the trends in tier 1 or 2 Ebefs

```python
%matplotlib inline
import pandas as pd
pd.options.display.mpl_style = 'default'
```

```python
data_df = pd.read_csv('../Data/ebef_aic_cross_ref.csv',sep = ',', parse_dates=["Created On"], 
                      dayfirst=True,index_col =["Created On"] )
```

```python
# Choose tier of interest
toi = 'AIC Part 1'
```

```python
# Group by tier
data_df_group = data_df.groupby(toi)
```

```python
# Plot the time series for the occurence of all types of Ebefs, on a weekly basis (starting Saturday)
import re
for key, grp in data_df_group:
    grp_quant = grp['Quantity']
    normalise()
    grp_weeks = grp_quant.resample('W-SAT', how='sum') # week
    grp_weeks[grp_weeks.isnull()] = 0
    
    # Don't plot topics with to few weekly reports 
    if grp_weeks.mean() < 10: continue
    ax = grp_weeks.plot(grid=None, label = key,figsize=(10,5))
    ax.set_yscale('log')
    fig = ax.get_figure()
    ax.set_ylabel('Frequency')
    #fig.savefig('../Outputs/%s.png' % re.sub('/','',key))

    # plt.legend(loc=0)
fig.subplots_adjust(bottom = 0.2)
fig.savefig('../Outputs/all.png' )
#     plt.close('all')
```
